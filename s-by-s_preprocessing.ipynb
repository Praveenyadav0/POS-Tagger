```python
f = open(f"./s-by-s_train_processed.txt", "w+")
for folder in train_folder_list:
    file_list = os.listdir(f"./Train-corpus/{folder}")
    for file in file_list:
        filename = f"./Train-corpus/{folder}/{file}"
        print(filename)
        name = filename.split('.')[0]
        tree = ET.parse(filename)
        root = tree.getroot()
        count_words = 0
        count_sentences = 0

        for sentence in root.iter('s'):
            count_sentences += 1
            line = ""
            for each_word in sentence.getchildren():
                if each_word.tag=="w": #only take words from word tag
                    count_words += 1
                    word = each_word.text.strip()
                    tag = each_word.attrib['c5'].strip()
                    line += f"{word}_{tag} "
            line = line.strip() # removing the last space
            line += "\n"
            #print(line)
            try:
                f.write(line)
            except:
                continue
        print(count_sentences)
f.close()
```

```python
f = open(f"./s-by-s_test_processed.txt", "w+")
for folder in test_folder_list:
    file_list = os.listdir(f"./Test-corpus/{folder}")
    for file in file_list:
        filename = f"./Test-corpus/{folder}/{file}"
        print(filename)
        name = filename.split('.')[0]
        tree = ET.parse(filename)
        root = tree.getroot()
        count_words = 0
        count_sentences = 0

        for sentence in root.iter('s'):
            count_sentences += 1
            line = ""
            for each_word in sentence.getchildren():
                if each_word.tag=="w": #only take words from word tag
                    count_words += 1
                    word = each_word.text.strip()
                    line += f"{word}_{each_word.attrib['c5']} "
            line = line.strip()
            line += "\n"
            #print(line)
            try:
                f.write(line)
            except:
                f.write("")
        print(count_sentences)
f.close()
```
