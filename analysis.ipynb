```python
import numpy as np
import pprint
from collections import Counter
```

```python
word_tag_pairs = []
f = open(f"./Train-corpus/processed.txt", "r")
word_tag_pairs = f.readlines()
f.close()
print(len(word_tag_pairs))
```

```python
word_tag_pairs = [word_tag_pair[:-1] for word_tag_pair in word_tag_pairs]
unique_word_tag_pairs = Counter(word_tag_pairs).most_common()
all_words = [pair.split('_')[0] for pair in word_tag_pairs]
all_tags = [pair.split('_')[1] for pair in word_tag_pairs]
word_freq = Counter(all_words).most_common() # represent how many times each word appeared in dataset
tag_freq = Counter(all_tags).most_common()# represent how many times each tag appeared in dataset
```

```python
print(len(tag_freq))
tag_freq
```

```python
word_freq
```

```python
freq_for_each_tag = []
for tag_n_count in tag_freq:
    tag = tag_n_count[0]
    print(tag)
    words_for_this_tag = [word_tag_pair for word_tag_pair in unique_word_tag_pairs if tag == word_tag_pair[0].split('_')[1]]
    freq_for_each_tag.append(words_for_this_tag)
```
