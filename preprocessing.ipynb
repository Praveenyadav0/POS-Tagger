```python
import xml.etree.ElementTree as ET
import os
train_folder_list = os.listdir("./Train-corpus")
test_folder_list = os.listdir("./Test-corpus")
```

```python
f = open(f"./train_processed.txt", "w+")
for folder in train_folder_list:
    file_list = os.listdir(f"./Train-corpus/{folder}")
    for file in file_list:
        filename = f"./Train-corpus/{folder}/{file}"
        print(filename)
        name = filename.split('.')[0]
        tree = ET.parse(filename)
        root = tree.getroot()
        count = 0

        for each_word in root.iter('w'):
            count = count + 1
            word = each_word.text.strip()
            line = f"{word}_{each_word.attrib['c5']}\n"
            #print(line)
            try:
                f.write(line)
            except:
                f.write("")
        print(count)
f.close()
```
